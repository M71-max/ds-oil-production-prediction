```python
corr_df = pd.DataFrame(oil_df.corr()[ouput_col].sort_values(ascending=False)[1:])
corr_df.columns=['Corr']
corr_df
```

```python
 #exclude the elements with over 90% correlation with output
input_columns = corr_df[(abs(corr_df['Corr']) > 0.2) & (corr_df['Corr'] < 0.90)].index
```

```python
input_columns
```

Apply ML

```python
x = oil_df[input_columns]
y = oil_df[ouput_col]
```

y = f(x) + e

```python
train_length = 3000  # days
X_train, Y_train = x.iloc[:train_length], y[:train_length]

X_test, Y_test = x.iloc[train_length:], y[train_length:]
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
# scaler = MinMaxScaler()
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# linear regression
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
```

```python
lm.fit(X_train_scaled, Y_train)
```

```python
yp_train = lm.predict(X_train_scaled)
```

```python
plt.figure(figsize=(12, 4))

plt.scatter(oil_df.index[:train_length], Y_train, color='orange')
plt.plot(oil_df.index[:train_length], yp_train)

plt.grid()
```

```python
yp_test = lm.predict(X_test_scaled)
```

```python
plt.figure(figsize=(12, 4))

plt.scatter(oil_df.index[:train_length], Y_train, color='orange')
plt.plot(oil_df.index[:train_length], yp_train)

plt.scatter(oil_df.index[train_length:], Y_test, color='green')
plt.plot(oil_df.index[train_length:], yp_test, color='black')

plt.axvline(oil_df.index[train_length], color='red')

plt.grid()
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
```

```python
rf.fit(X_train, Y_train)
yp_train = rf.predict(X_train)
yp_test = rf.predict(X_test)
```

```python
plt.figure(figsize=(12, 4))

plt.scatter(oil_df.index[:train_length], Y_train, color='orange')
plt.plot(oil_df.index[:train_length], yp_train)

plt.scatter(oil_df.index[train_length:], Y_test, color='green')
plt.plot(oil_df.index[train_length:], yp_test, color='black')

plt.axvline(oil_df.index[train_length], color='red')

plt.grid()
```
